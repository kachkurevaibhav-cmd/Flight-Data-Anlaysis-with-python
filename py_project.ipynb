```python
#  showing all Airlines with their number of flights in bar graph

df['airline'].value_counts(ascending=True).plot.barh( color=['lightgreen','grey'])
plt.title("Airlines with frequencies")
plt.xlabel("Number of flights")
plt.ylabel("Airlines")
plt.show()
```

```python
# showing the departure time for the flights.

df['departure_time'].value_counts()
```

```python
# showing the arrival time for the flights.

df['arrival_time'].value_counts()
```

```python
plt.figure(figsize = (14,4))
plt.subplot(1,2,1)
plt.bar(df['departure_time'].value_counts().index ,df['departure_time'].value_counts().values, color =['red','gray'] )
plt.title("Department Time")
plt.xlabel("Department time")
plt.ylabel("department frequencies")



plt.subplot(1,2,2)
plt.bar(df['arrival_time'].value_counts().index,df['arrival_time'].value_counts().values, color = ['g','y'])
plt.title('Arrial Time ')
plt.xlabel('A time ')
plt.ylabel('A freq')
plt.show()
```

```python
# source city and Destination city for the flight counts

plt.figure(figsize = ( 16,4))
plt.subplot(1,2,1)
plt.barh(df['source_city'].value_counts().index,df['source_city'].value_counts().values,color = ['m','g'])
plt.title(" source city with no of flights")
plt.xlabel("Cities")
plt.ylabel("no of flights")





plt.subplot(1,2,2)
plt.bar(df['destination_city'].value_counts().index,df['destination_city'].value_counts().values,color = ['m','g'])
plt.title("destination_city  with no of flights")
plt.xlabel("Cities")
plt.ylabel("no of flights")


plt.show()
```
